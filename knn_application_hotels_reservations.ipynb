```python
from sklearn.neighbors import KNeighborsClassifier
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
project_path = "D:\\viks\\class_work\\class_work_anurag"
file_path = project_path + '\\Hotel_Reservations.csv'

df = pd.read_csv(file_path)
print(df)
```

```python
inp = df.drop('booking_status',axis=1)       
print(inp)

out = df['booking_status']
print(out)
```

```python
train_x, test_x, train_y, test_y = train_test_split(inp, out, test_size = 0.3)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_x, train_y)
```

```python
y_pred = knn.predict(test_x)

print(knn.predict(test_x))
print(y_pred[0:5])
# Calculate the accuracy of the model
print(knn.score(test_x, test_y))
```

```python
for k in range(3,11,2):
    train_x, test_x, train_y, test_y = train_test_split(inp, out, test_size = 0.1)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(train_x, train_y)
    y_pred = knn.predict(test_x)

    #print(knn.predict(test_x))
    # Calculate the accuracy of the model
    print('k=',k,'score=',knn.score(test_x, test_y))
    
```
